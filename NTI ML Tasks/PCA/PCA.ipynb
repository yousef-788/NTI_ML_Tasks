### Import Liabraries

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
```

### Import Data

```python
data = pd.read_csv('HR_comma_sep.csv')
```

### Data Encoding

```python
data = pd.get_dummies(data, columns=['salary', 'sales'], drop_first=True)
```

### Slpitting

```python
X = data.drop('left', axis=1)
y = data['left']
```

### Standard Normalization

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

# Model

### Without PCA

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
y_pred = rf_model.predict(X_test)


acc_before_pca = accuracy_score(y_test, y_pred)
print(f" Accuracy BEFORE PCA: {acc_before_pca:.4f}")
```

### With PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
Xpca = pca.fit_transform(X_scaled)
```

```python
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(Xpca, y, test_size=0.3, random_state=42)
```

```python
clf = RandomForestClassifier(random_state=42)
clf.fit(X_train_pca, y_train_pca)
```

```python

y_pred_pca = clf.predict(X_test_pca)
acc_after_pca = accuracy_score(y_test_pca, y_pred_pca)
print(f" Accuracy AFTER PCA: {acc_after_pca:.4f}")
```
